```python
def nearest(row, geom_union, df1, df2, geom1_col='geometry', geom2_col='geometry', src_column=None):
    """Find the nearest point and return the corresponding value from specified column."""
    # Find the geometry that is closest
    nearest = df2[geom2_col] == nearest_points(row[geom1_col], geom_union)[1]
    # Get the corresponding value from df2 (matching is based on the geometry)
    value = df2[nearest][src_column].get_values()[0]
    return value
```

```python
lines_df = all_kml[['ROUTE_NAME']].drop_duplicates().sort_values(by = ['ROUTE_NAME']).reset_index(drop=True)

from tqdm import tqdm_notebook as tqdm
all_routes_track = gpd.GeoDataFrame()
for i in tqdm(range(len(lines_df))):
    ROUTE_NAME = lines_df.ROUTE_NAME[i]
    one_route = all_kml.query('ROUTE_NAME == @ROUTE_NAME').reset_index(drop=True)
    one_track = one_route[['nod_id','name','route_date_name','ROUTE_NAME','geometry']].query('name != 1').reset_index().rename(columns={'index':'stop_seq'})
    ROUTE_NAME = str(ROUTE_NAME)
    one_stop_sequence = stop_sequence.query('ROUTE_NAME == @ROUTE_NAME').reset_index(drop=True)
    unary_union = one_stop_sequence.unary_union
    one_track['nearest_id'] = one_track.apply(nearest, geom_union=unary_union, df1=one_track, df2=one_stop_sequence, geom1_col='geometry', src_column='STOP_ID', axis=1)
    one_route = one_route.merge(one_track[['nod_id','route_date_name','nearest_id','stop_seq']], how = 'left', on = ['nod_id','route_date_name']) 
    one_route = one_route.fillna(-1)
    one_route = one_route.rename(columns = {'nearest_id':'stop_id'})
    one_route['stop_id'] = one_route['stop_id'].astype(np.int64)
    trips = one_route[['trip_num']].drop_duplicates().reset_index(drop=True)
    routes_trips = pd.DataFrame()
    for j in range(len(trips)):
        trip = trips.trip_num[j]
        one_trip = one_route.query('trip_num == @trip').reset_index(drop=True)
        list_stop_group = []
        list_stop_group.append(one_trip.stop_id[0])
        list_seq_group = []
        list_seq_group.append(one_trip.stop_seq[0])
        for x in range(1,len(one_trip)):    
            if one_trip.stop_id[x] == -1:
                list_stop_group.append(list_stop_group[-1])
                list_seq_group.append(list_seq_group[-1])
            else:
                list_stop_group.append(one_trip.stop_id[x])
                list_seq_group.append(one_trip.stop_seq[x])
        one_trip['main_stop_id'] = list_stop_group
        one_trip['main_stop_seq'] = list_seq_group
        routes_trips = routes_trips.append(one_trip)
    routes_trips = routes_trips.reset_index(drop = True)
    matrix_com = pd.merge(routes_trips, routes_trips, how='inner', on=['trip_num'])
    matrix_com = matrix_com.sort_values(by = ['trip_num','nod_id_x', 'nod_id_y'])
    matrix_com = matrix_com.rename(columns={'nod_id_x':'start_nod_id', 'geometry_x':'start_geometry',
                                            'nod_id_y':'end_nod_id', 'geometry_y':'end_geometry'})
    matrix_com = matrix_com.query('start_nod_id < end_nod_id')
    matrix_stages = matrix_com[matrix_com.end_nod_id == (matrix_com.start_nod_id + 1)]
    matrix_stages = matrix_stages.reset_index(drop=True)
    matrix_stages['long'] = 3600*(matrix_stages['hour_y'] - matrix_stages['hour_x']) + 60*(matrix_stages['minuts_y'] - matrix_stages['minuts_x']) + (matrix_stages['sek_y'] - matrix_stages['sek_x'])
    matrix_stages['length'] = np.sqrt((matrix_stages['lon_y']-matrix_stages['lon_x'])**2 + (matrix_stages['lat_y']-matrix_stages['lat_x'])**2)
    matrix_stages['speed'] = ((matrix_stages['length'])/(matrix_stages['long']))*3.6
    matrix_stages = matrix_stages[['trip_num', 'start_nod_id','main_stop_id_x','main_stop_seq_x','start_geometry','time_x','end_nod_id','main_stop_id_y','main_stop_seq_y', 'end_geometry','time_y','long','length','speed']]
    matrix_stages = matrix_stages.replace([np.inf, -np.inf], np.nan).fillna(0)
    
    means_and_sum_1 = matrix_stages.groupby(['main_stop_seq_x'])['long'].agg(['sum']).reset_index()
    means_and_sum_1 = means_and_sum_1.rename(columns={'sum':'sum_seg_long'})
    means_and_sum_2 = matrix_stages.groupby(['main_stop_seq_x'])['length'].agg(['sum']).reset_index()
    means_and_sum_2 = means_and_sum_2.rename(columns={'sum':'sum_seg_length'})
    means_and_sum = means_and_sum_1.merge(means_and_sum_2, how='left', on=['main_stop_seq_x'])
    means_and_sum['mean_seg_speed'] = ((means_and_sum['sum_seg_length'])/(means_and_sum['sum_seg_long']))*3.6
    means_and_sum = means_and_sum.replace([np.inf, -np.inf], np.nan).fillna(0)

    means_and_sum_3 = matrix_stages.groupby(['trip_num'])['long'].agg(['sum']).reset_index()
    means_and_sum_3 = means_and_sum_3.rename(columns={'sum':'sum_long'})
    means_and_sum_4 = matrix_stages.groupby(['trip_num'])['length'].agg(['sum']).reset_index()
    means_and_sum_4 = means_and_sum_4.rename(columns={'sum':'sum_length'})
    means_and_sum_5 = means_and_sum_3.merge(means_and_sum_4, how='left', on=['trip_num'])
    means_and_sum_5['mean_speed'] = ((means_and_sum_5['sum_length'])/(means_and_sum_5['sum_long']))*3.6
    means_and_sum_5 = means_and_sum_5.replace([np.inf, -np.inf], np.nan).fillna(0)

    matrix_stages = matrix_stages.merge(means_and_sum, how='left', on=['main_stop_seq_x'])
    matrix_stages = matrix_stages.merge(means_and_sum_5, how='left', on=['trip_num'])
    
    
    all_routes_track=all_routes_track.append(matrix_stages)
    
all_routes_track = all_routes_track.reset_index(drop=True)
all_routes_track
```
