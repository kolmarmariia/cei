```python
import pandas as pd
f = pd.read_csv('parametres.csv', sep=';', encoding='windows-1251')
```

```python
filtr = f[f.MARSH == 'Тм1']
```

```python
names = f.MARSH
```

```python
import docx
from docx.enum.section import WD_ORIENT
from docx.enum.table import WD_ALIGN_VERTICAL
from docx.enum.table import WD_TABLE_ALIGNMENT
from docx.enum.style import WD_STYLE
from docx.dml.color import ColorFormat
from docx.enum.dml import MSO_COLOR_TYPE
from docx.enum.text import WD_COLOR_INDEX
from docx.enum.text import WD_COLOR
```

```python
doc = docx.Document()
section = doc.sections[-1]
new_width, new_height = section.page_height, section.page_width
section.orientation = WD_ORIENT.LANDSCAPE
section.page_width = new_width
section.page_height = new_height
x=0
for k in names:
    x=x+1
    file_name=str(k)
    file_name2=file_name.replace("Тм", "Tm").replace("А", "A").replace("Тб", "Tb").replace("Эб", "A")
    filtr = f[f.MARSH == k]
    doc.add_picture('./Копии/'+file_name2+'.png', width = docx.shared.Cm(18))
    table = doc.add_table(rows = 3, cols = 10)
    table.style = 'Table Grid'
        # получаем ячейку таблицы
    cell = table.cell(0,0)
    cell.text = 'Маршрут'
    cell2 = table.cell(0, 1)
    cell3 = table.cell(1, 0) 
    cell4 = table.cell(1, 1)
    cell2.width = 3298280 
    cell4.width = 3298280 
    cell.merge(cell2)
    cell3.merge(cell4)
    cell.merge(cell3)
    cell5 = table.cell(0, 2)
    cell5.text = 'Класс ТС'
    cell6 = table.cell(0, 3)
    cell7 = table.cell(0, 4)
    cell8 = table.cell(0, 5)
    cell5.merge(cell6)
    cell5.merge(cell7)
    cell5.merge(cell8)
    cell9 = table.cell(0, 6)
    cell9.text = 'Интервалы'
    cell10 = table.cell(0, 7)
    cell11 = table.cell(0, 8)
    cell12 = table.cell(0, 9)
    cell9.merge(cell10)
    cell9.merge(cell11)
    cell9.merge(cell12)
    cell13 = table.cell(1, 2)
    cell13.text = 'МК'
    cell14 = table.cell(1, 3)
    cell14.text = 'СК'
    cell15 = table.cell(1, 4)
    cell15.text = 'БК'
    cell16 = table.cell(1, 5)
    cell16.text = 'ОБК'
    cell17 = table.cell(1, 6)
    cell17.text = 'утр.пик'
    cell18 = table.cell(1, 7)
    cell18.text = 'м/пик'
    cell19 = table.cell(1, 8)
    cell19.text = 'веч.пик'
    cell20 = table.cell(1, 9)
    cell20.text = 'после 21.00'
    # заполняем таблицу данными
    
    cell21 = table.cell(2,0)
    cell21.text = filtr['MARSH']
    lol = filtr['Маршрут,Начальный пункт'] + '-' + filtr['Маршрут,Конечный пункт']
    cell22 = table.cell(2,1)
    cell22.text = lol
    cell23 = table.cell(2,2)
    cell23.text = str(filtr['Выпуск в час пик по классам вместимости,МК (Тм БК)']).split('\n')[0].split(' ')[-1].replace("NaN", "").replace(".0", "")
    cell24 = table.cell(2,3)
    cell24.text = str(filtr['Выпуск в час пик по классам вместимости,СК(Тм ОБК)']).split('\n')[0].split(' ')[-1].replace("NaN", "").replace(".0", "")
    cell25 = table.cell(2,4)
    cell25.text = str(filtr['Выпуск в час пик по классам вместимости,БК (Тм 2хБВ)']).split('\n')[0].split(' ')[-1].replace("NaN", "").replace(".0", "")
    cell26 = table.cell(2,5)
    cell26.text = str(filtr['Выпуск в час пик по классам вместимости,ОБК(Тм 3хБВ)']).split('\n')[0].split(' ')[-1].replace("NaN", "").replace(".0", "")
    cell27 = table.cell(2,6)
    cell27.text = str(filtr['Интервал будни,Ут.Пик']).split('\n')[0].split(' ')[-1].replace("NaN", "")
    cell28 = table.cell(2,7)
    cell28.text = str(filtr['Интервал будни,День']).split('\n')[0].split(' ')[-1].replace("NaN", "")
    cell29 = table.cell(2,8)
    cell29.text = str(filtr['Интервал будни,Веч.пик']).split('\n')[0].split(' ')[-1].replace("NaN", "")
    cell30 = table.cell(2,9)
    cell30.text = str(filtr['Интервал будни,Ночь']).split('\n')[0].split(' ')[-1].replace("NaN", "")
    
    for row in range(3):
          for col in range(10):
                table.cell(row, col).paragraphs[0].alignment  = WD_ALIGN_VERTICAL.CENTER
    doc.add_page_break()
section = doc.sections[-1]
section.orientation = WD_ORIENT.LANDSCAPE
doc.save("res.docx".format(id=k))
```
